```python
# Importing 2021 datasets

usa21 = pd.read_excel("USA_2021.xlsx", sheet_name='Data')
sf21 = pd.read_excel("SF_2021.xlsx", sheet_name='Data')
sd21 = pd.read_excel("San Diego 2021.xlsx", sheet_name='Data')
```

```python
usa21.head()
```

```python
# Functions to clean 2021 datasets

def clean21(df):

  """ This function extracts only rows of "All establishments"
      and removes unnecessary columns """

  df = df[(df['Meaning of Employment size of establishments code'] == 'All establishments') & \
          (df['Meaning of Legal form of organization code'] == 'All establishments')]

  dropCols = ['Number of establishments', 'Annual payroll ($1,000)', 'First-quarter payroll ($1,000)']
  df.drop(columns=dropCols, inplace=True)

  replaceDict = {'N': 0}
  df['Number of employees'] = df['Number of employees'].replace(replaceDict)

  return df
```

```python
usa21 = clean21(usa21)
sf21 = clean21(sf21)
sd21 = clean21(sd21)
```

```python
# Importing USA 2012 dataset

keepCols = ['NAME', 'NAICS2012', 'NAICS2012_LABEL', 'LFO_LABEL', 'EMPSZES_LABEL', 'YEAR', 'EMP']

usa12 = pd.read_csv("USA_2012.csv", usecols=keepCols)
usa12.head()
```

```python
usa12.head()
```

```python
usa12.drop([0], axis=0, inplace=True)
```

```python
# Function to clean EMP column of 2012 US dataset

def clean12(val):
  parts = val.split(' ')
  num = []
  for part in parts:
    part = part.replace(',', '')
    try:
      part = float(part)
      num.append(part)
    except:
      continue
  return np.mean(num)
```

```python
usa12['EMP'] = [clean12(val) for val in usa12['EMP']]
```

```python
# sanity check
np.unique(usa12[usa12['NAICS2012'] == '236']['NAICS2012_LABEL'])
```

```python
# Get San Francisco County data for 2012

sf12 = usa12[usa12['NAME'].str.contains('San Francisco')]
```

```python
sf12.reset_index(drop=True, inplace=True)
```

```python
sf12.head()
```

```python
# Get San Diego County data for 2012

sd12 = usa12[usa12['NAME'].str.contains('San Diego')]
```

```python
sd12.reset_index(drop=True, inplace=True)
```

```python
sd12.head()
```

```python
# Convert NAICS2012 code column to string

usa12['NAICS2012'] = usa12['NAICS2012'].astype(str)
usa12['NAICS2012_LABEL'] = usa12['NAICS2012_LABEL'].astype(str)
usa12['YEAR'] = usa12['YEAR'].astype(str)
```

```python
# For USA 2012, get total employment (EMP) by NAICS code

usa12Final = usa12.groupby(by='NAICS2012').agg({'NAICS2012_LABEL': min, 'YEAR': min, 'EMP': sum}).reset_index()
```

```python
# Miscellaneous functions

def export_dataset(df, filename):

  """ This function export dataframe as csv file """

  df.to_csv(filename)
  print(f'Successfully saved {filename}!')
```

```python
export_dataset(usa21, 'usa21.csv')
export_dataset(usa12Final, 'usa12.csv')
export_dataset(sf21, 'sf21.csv')
export_dataset(sf12, 'sf12.csv')
export_dataset(sd21, 'sd21.csv')
export_dataset(sd12, 'sd12.csv')
```
